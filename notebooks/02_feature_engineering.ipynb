```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/ResaleflatpricesbasedonregistrationdatefromJan2017onwards.csv')
print(df.shape)
df.head()
```

```python
# 1: Extract year and month from transaction date
df['year'] = df['month'].str[:4].astype(int)
df['transaction_month'] = df['month'].str[5:7].astype(int)

print(df[['month', 'year', 'transaction_month']].head(10))
```

```python
# 2: Convert storey_range to midpoint number
def extract_storey_midpoint(storey_range):
    parts = storey_range.split(' TO ')
    low = int(parts[0])
    high = int(parts[1])
    return (low + high) / 2

df['storey_midpoint'] = df['storey_range'].apply(extract_storey_midpoint)

print(df[['storey_range', 'storey_midpoint']].head(10))
```

```python
# 3: Convert remaining_lease text to decimal years
def parse_remaining_lease(lease_str):
    parts = lease_str.split(' ')
    years = int(parts[0])
    
    # some entries only have years, no months
    if 'month' in lease_str:
        months = int(parts[2])
    else:
        months = 0
        
    return years + (months / 12)

df['remaining_lease_years'] = df['remaining_lease'].apply(parse_remaining_lease)

print(df[['remaining_lease', 'remaining_lease_years']].head(10))
```

```python
# 4: Encode flat_type as ordered integer
flat_type_order = {
    '1 ROOM': 1,
    '2 ROOM': 2,
    '3 ROOM': 3,
    '4 ROOM': 4,
    '5 ROOM': 5,
    'EXECUTIVE': 6,
    'MULTI-GENERATION': 7
}

df['flat_type_encoded'] = df['flat_type'].map(flat_type_order)

print(df[['flat_type', 'flat_type_encoded']].value_counts())
```

```python
# 5: One-hot encode town
town_dummies = pd.get_dummies(df['town'], prefix='town')
print(f"Number of town columns created: {len(town_dummies.columns)}")
print(town_dummies.columns.tolist())
print(town_dummies.head())
```

```python
# 6: One-hot encode flat_model
flat_model_dummies = pd.get_dummies(df['flat_model'], prefix='flat_model')
print(f"Number of flat_model columns created: {len(flat_model_dummies.columns)}")
print(flat_model_dummies.columns.tolist())
```

```python
# Assemble final feature DataFrame
df_features = pd.DataFrame({
    # Original numerical features
    'floor_area_sqm': df['floor_area_sqm'],
    'lease_commence_date': df['lease_commence_date'],
    
    # Engineered features
    'year': df['year'],
    'transaction_month': df['transaction_month'],
    'storey_midpoint': df['storey_midpoint'],
    'remaining_lease_years': df['remaining_lease_years'],
    'flat_type_encoded': df['flat_type_encoded'],
})

# Add one-hot encoded columns
df_features = pd.concat([df_features, town_dummies, flat_model_dummies], axis=1)

# Add target variable
df_features['resale_price'] = df['resale_price']

print("Final feature set shape:", df_features.shape)
print("\nColumns:")
print(df_features.columns.tolist())
print("\nFirst few rows:")
df_features.head()
```

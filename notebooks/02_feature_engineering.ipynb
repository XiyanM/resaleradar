# Feature Engineering
This notebook transforms raw HDB resale transaction data into a dataset suitable for XGBoost training, with each section below addressing a specific transformation.

## Setup
Import required libraries and load the raw dataset.


```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/ResaleflatpricesbasedonregistrationdatefromJan2017onwards.csv')
print(df.shape)
df.head()
```

## Feature 1: Transaction Date Extraction
The raw `month` column stores dates in `YYYY-MM` string format (e.g. `2017-01`).

We extract `year` and `transaction-month` as separate integer features because:

- `year` captures long-term price trends and market cycles
- `transaction_month` captures seasonal patterns in resale activity
- Models require numerical inputs — raw date strings cannot be used directly

```python
df['year'] = df['month'].str[:4].astype(int)
df['transaction_month'] = df['month'].str[5:7].astype(int)

print(df[['month', 'year', 'transaction_month']].head(10))
```

## Feature 2: Storey Range to Midpoint

The raw `storey_range` column contains text ranges like `"07 TO 09"`.

We extract a single numerical midpoint value (e.g 8.0) because:

- Higher floors command price premiums in the Singapore HDB market
- The midpoint preserves the ordinal relationship in a format the model can learn from
- Models require numerical inputs — text ranges cannot be used directly

```python
def extract_storey_midpoint(storey_range):
    parts = storey_range.split(' TO ')
    low = int(parts[0])
    high = int(parts[1])
    return (low + high) / 2

df['storey_midpoint'] = df['storey_range'].apply(extract_storey_midpoint)

print(df[['storey_range', 'storey_midpoint']].head(10))
```

## Feature 3: Remaining Lease Duration

The raw `remaining_lease` column stores duration as text in three formats:
- `"61 years 04 months"`
- `"62 years 01 month"`
- `"63 years"`

We extract a single decimal value representing total years (e.g. 61.33) because:

- Remaining lease is one of the strongest predictors of HDB resale price
- Flats with shorter leases trade at significant discounts due to CPF restrictions
- Models require a single numerical input, not free-form text

```python
def parse_remaining_lease(lease_str):
    parts = lease_str.split(' ')
    years = int(parts[0])
    
    # some entries only have years, no months
    if 'month' in lease_str:
        months = int(parts[2])
    else:
        months = 0
        
    return years + (months / 12)

df['remaining_lease_years'] = df['remaining_lease'].apply(parse_remaining_lease)

print(df[['remaining_lease', 'remaining_lease_years']].head(10))
```

## Feature 4: Flat Type Encoding

The `flat_type` column contains 7 categories with a natural size ordering:
`1 ROOM < 2 ROOM < 3 ROOM < 4 ROOM < 5 ROOM < EXECUTIVE < MULTI-GENERATION`

We use label encoding (integers 1–7) rather than one-hot encoding because:

- The ordering is meaningful — the model should know a 5 ROOM is "more" than a 3 ROOM
- One-hot encoding would lose this ordinal information
- Label encoding keeps it as a single column, reducing dimensionality

```python
flat_type_order = {
    '1 ROOM': 1,
    '2 ROOM': 2,
    '3 ROOM': 3,
    '4 ROOM': 4,
    '5 ROOM': 5,
    'EXECUTIVE': 6,
    'MULTI-GENERATION': 7
}

df['flat_type_encoded'] = df['flat_type'].map(flat_type_order)

print(df[['flat_type', 'flat_type_encoded']].value_counts())
```

## Feature 5: Town One-Hot Encoding

The raw `town` column contains 26 unique towns with no natural ordering.

We use one-hot encoding via `pd.get_dummies()` because:

- There is no meaningful order between towns (BISHAN is not "greater than" BEDOK)
- Label encoding would imply a false numerical relationship between towns
- One-hot encoding treats each town as an independent binary feature

Note: We retain all 26 columns despite the dummy variable trap because XGBoost 
uses tree-based splits rather than matrix operations, so multicollinearity is not an issue.

```python
# 5: One-hot encode town
town_dummies = pd.get_dummies(df['town'], prefix='town')
print(f"Number of town columns created: {len(town_dummies.columns)}")
print(town_dummies.columns.tolist())
print(town_dummies.head())
```

## Feature 6: Flat Model One-Hot Encoding

The `flat_model` column contains 21 unique categories with no natural ordering 
(e.g. Improved, New Generation, DBSS, Premium Apartment).

We apply the same one-hot encoding approach as `town`, producing 21 binary columns.

```python
flat_model_dummies = pd.get_dummies(df['flat_model'], prefix='flat_model')
print(f"Number of flat_model columns created: {len(flat_model_dummies.columns)}")
print(flat_model_dummies.columns.tolist())
```

## Feature Assembly

We combine all engineered features into a single DataFrame using `pd.concat()`.

Final feature set:
- 2 original numerical features: `floor_area_sqm`, `lease_commence_date`
- 5 engineered numerical features: `year`, `transaction_month`, `storey_midpoint`, `remaining_lease_years`, `flat_type_encoded`
- 26 town dummy columns
- 21 flat model dummy columns
- 1 target variable: `resale_price`

**Total: 55 columns, 233,055 rows**

Columns excluded:
- `month` → replaced by `year` and `transaction_month`
- `block`, `street_name` → too high cardinality, requires geocoding to be useful
- `storey_range` → replaced by `storey_midpoint`
- `remaining_lease` → replaced by `remaining_lease_years`
- `flat_type`, `town`, `flat_model` → replaced by encoded versions

```python
df_features = pd.DataFrame({
    'floor_area_sqm': df['floor_area_sqm'],
    'lease_commence_date': df['lease_commence_date'],
    
    'year': df['year'],
    'transaction_month': df['transaction_month'],
    'storey_midpoint': df['storey_midpoint'],
    'remaining_lease_years': df['remaining_lease_years'],
    'flat_type_encoded': df['flat_type_encoded'],
})

df_features = pd.concat([df_features, town_dummies, flat_model_dummies], axis=1)

df_features['resale_price'] = df['resale_price']

print("Final feature set shape:", df_features.shape)
print("\nColumns:")
print(df_features.columns.tolist())
print("\nFirst few rows:")
df_features.head()
```

## Save Processed Dataset

We save the final feature set to `data/processed/features.csv`.

This allows the training notebook to read from the processed data directly instead of re-running all feature engineering steps during training.



```python
df_features.to_csv('../data/processed/features.csv', index=False)
print(f"Saved to data/processed/features.csv")
print(f"Shape: {df_features.shape}")
```
